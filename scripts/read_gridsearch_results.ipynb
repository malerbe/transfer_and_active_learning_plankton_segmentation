```python
import json

logs_dir = r"...\logs"

with open(logs_dir + r'\GridSearch_10\grid_search_results.json', 'r') as file:
    content = json.load(file)

print(content)
best_test_loss, best_test_loss_run_id = None, None
best_dice, best_dice_run_id = None, None
best_iou, best_iou_run_id = None, None
for run in content:
    test_loss, dice, iou = run['best_metrics']['test_loss'], run['best_metrics']['test_dice'], run['best_metrics']['test_iou']

    if best_test_loss is None or test_loss > best_test_loss:
        best_test_loss = test_loss
        best_test_loss_run_id = run['run_id']

    if best_dice is None or dice > best_dice:
        best_dice = dice
        best_dice_run_id = run['run_id']

    if best_test_loss is None or iou > best_test_loss:
        best_iou = iou
        best_iou_run_id = run['run_id']

print("="*20)
print("Test Loss")
print(f"Best run : {best_test_loss_run_id} ; Test Loss : {best_test_loss}")

print("="*20)
print("Dice")
print(f"Best run : {best_dice_run_id} ; Test Dice : {best_dice}")

print("="*20)
print("IoU")
print(f"Best run : {best_iou_run_id} ; Test IoU : {best_iou}")
```
